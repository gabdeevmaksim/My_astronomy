```python
from astropy.io import fits
import numpy as np
import pandas as pd
```

```python
fit= fits.open('R0605480101M1S001SRCTSR0001.FIT')
```

```python
fit.info()
```

```python
lc = fit[1].data
```

```python
lc_pd = pd.DataFrame(lc)
```

```python
import matplotlib.pyplot as plt
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
plt.plot(lc_pd.TIME-lc_pd.TIME[0],smooth(lc_pd.RATE,3))
plt.xlim(15000,30000)
```
