```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
m45 = pd.read_csv('e:\Observations\MMPP\\20201013\Mags_m45_grw.csv')
p45 = pd.read_csv('e:\Observations\MMPP\\20201013\Mags_p45_grw.csv')
```

```python
u = (10.**(m45.iloc[:,5]-m45.iloc[:,5+1])-1.)/(10.**(m45.iloc[:,5]-m45.iloc[:,5+1])+1.)
q = (10.**(p45.iloc[:,5]-p45.iloc[:,5+1])-1.)/(10.**(p45.iloc[:,5]-p45.iloc[:,5+1])+1.)
(u-q)*50
```

```python
N=m45.shape[1]-2
Pol = np.empty([14,40])
Mag = np.empty([14,40])
for i in list(range(2,27,2)):
    u = (10.**(m45.iloc[:,i]-m45.iloc[:,i+1])-1.)/(10.**(m45.iloc[:,i]-m45.iloc[:,i+1])+1.)
    q = (10.**(p45.iloc[:,i]-p45.iloc[:,i+1])-1.)/(10.**(p45.iloc[:,i]-p45.iloc[:,i+1])+1.) 
    m = (m45.iloc[:,i]+m45.iloc[:,i+1])/2.
    p = np.array(50*(u-q)) 
    Mag[int(i/2-1)][:] = m 
    Pol[int(i/2-1)][:] = p                                                   
```

```python
fig01 = plt.figure(1,figsize=[8,8])
#plt.hist(np.median(Pol, axis=1))
#plt.hist(np.std(Pol, axis=1))
plt.semilogx(18000/10**(0.4*np.median(Mag,axis=1)),np.std(Pol,axis=1), 'bo')
#plt.scatter(np.median(Mag,axis=1),np.std(Pol,axis=1))
plt.rc('font', size=15) 
#plt.rc('axes', titlesize=12)
plt.xlabel('Flux in ordinary light')
plt.ylabel('$\sigma$, %')
plt.savefig('e:\Observations\MMPP\\20201013\\fig01.eps')
```
