```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = path = 'C:\\Users\\gamak\\Documents\\Papers\\CVs in eROS\\'
tab = pd.read_csv(path+'gaia_xstars_pm_r98_200pc_lkh10_woutdup.gzip')
```

```python
d = tab.dist_pc <= 50 #100
tab1 = tab[d]
```

```python
i = tab1.Teff < 3500
t1 = tab1[i]
m = t1.Teff >= 2000
j = tab1.Teff < 6000
t2 = tab1[j]
f = t2.Teff >= 5000 
```

```python
print(m.sum(),f.sum())
```

```python
fig = plt.figure(figsize=(7,7))
plt.plot(t1.dist_pc[m],np.log10(t1.Lx[m]),'ro', markersize=1)
plt.plot(t2.dist_pc[f],np.log10(t2.Lx[f]),'go', markersize=1)
plt.xlabel('Distance, pc')
plt.ylabel('LogLx')
plt.title('M(red) and F(green) stars X-ray luminosities in 100 pc')
plt.savefig(path+'M&F_100pc.jpeg', dpi = 300)
```
