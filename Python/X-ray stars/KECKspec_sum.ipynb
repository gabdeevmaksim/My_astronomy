```python
import pandas as pd
import matplotlib.pyplot as plt

DIR = 'e:\\Observations\\KECK1\\SRGe0339+3722\\'
files = ['lris20220126_SeJ0339+3722_exp1.spec','lris20220126_SeJ0339+3722_exp2.spec','lris20220126_SeJ0339+3722_exp2.spec']

spec1 = pd.read_fwf(DIR+files[0], header = 137, widths = (2,10,13,13,13,8,10,13,8))
spec2 = pd.read_fwf(DIR+files[1], header = 137, widths = (2,10,13,13,13,8,10,13,8))
spec3 = pd.read_fwf(DIR+files[2], header = 137, widths = (2,10,13,13,13,8,10,13,8))
```

```python
s = (spec1.flux+spec2.flux)+spec3.flux)
```
