```python
import pandas as pd
from astroquery.vizier import Vizier
from astropy.coordinates import SkyCoord
from astropy import units as u 
from astropy.table import Table
```

```python
Vizier.ROW_LIMIT = -1
result = Vizier.get_catalogs('J/A+A/616/A10/tablea1a')
```

```python
i = result[0]['Cluster'] == 'Pleiades'
```

```python
table = result[0]
```

```python
pd_tab = table.to_pandas()
```

```python
pd_tab['Index'] = pd_tab.index
```

```python
pd_tab[['Source','RAdeg','DEdeg','Gmag','plx','e_plx']][i].to_csv('Pleiades_GAIA.csv', index=True)
```
