```python
from astropy.io import fits
import os
import matplotlib.pyplot as plt
import pandas as pd

os.chdir('e:/Observations/KECK1/')
```

```python
tab = fits.getdata('keck_lris_red_400.fits', ext=1)
```

```python
plt.plot(tab['wave'],tab['flux'])
```

```python
from scipy.signal import find_peaks
```

```python
fig = plt.figure(figsize = (25,10))
peaks, _ = find_peaks(tab['flux'], height=10000)
plt.plot(tab['wave'],tab['flux'])
for peak in peaks:
    plt.text(tab['wave'][peak], tab['flux'][peak], str(round(tab['wave'][peak],2)), rotation = 'vertical')
plt.xlim(5800,10000)
    
```
